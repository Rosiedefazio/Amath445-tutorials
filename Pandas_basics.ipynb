```python
from sklearn.datasets import load_iris

df = load_iris(as_frame=True).frame
target = df.pop("target")                   # remove the target column from the dataframe

# randomly add missing values to the dataframe
df = df.mask(np.random.random(df.shape) < 0.1)
print(df.head())

#Check for missing values
print(df.isnull().sum())

#Drop rows with missing valiues and place it in a new variable "df_cleaned"
df_cleaned = df.dropna()
print(df_cleaned.head())
```

```python
# Replace missing values in a specific column:
df['sepal length (cm)'] = df['sepal length (cm)'].fillna(df['sepal length (cm)'].mean())
```

```python
df.isnull().sum()
```

Create one-hot encoding for categorical columns using `pd.get_dummies()`:

```python
target_categorical = pd.get_dummies(target)
target_categorical
```
